```python
def PATTERNCOUNT(Text, Pattern):
    count = 0
    for i in range(0,len(Text) - len(Pattern) + 1):
        if Text[i : i + len(Pattern)] == Pattern:
            count = count + 1
    return count

print(PATTERNCOUNT("CGATATATCCATAG","ATA"))
```

```python
def freqwords(Text, k):
  freqpatterns = []
  countarray = []

  for i in range(0, len(Text) - k + 1):
    pattern = Text[i : i + k]
    countarray.append(PATTERNCOUNT(Text, pattern))

  if not countarray: # Handle case where Text is shorter than k
    return []

  maxcount = max(countarray)

  for i in range(0, len(Text) - k + 1):
    if countarray[i] == maxcount:
      freqpatterns.append(Text[i : i + k])

  # Remove duplicates from freqpatterns
  return list(set(freqpatterns))


print(freqwords('ACTGACT', 3))
```

```python
def count_kmers_fast(text, k):
    """
    Counts all k-mers in a text efficiently using a dictionary.
    """
    counts = {}
    for i in range(len(text) - k + 1):
        kmer = text[i:i+k]
        counts[kmer] = counts.get(kmer, 0) + 1
    return counts

kmer_frequencies = count_kmers_fast(genome, 3)
print(kmer_frequencies)
```
